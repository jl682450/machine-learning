<h2>Data Pre-processing:</h2>
<h4>W.R.T Weather Data</h4>
<p> 1. Built and calling reusable function 1 for cleaning and processing the data such as index setting, removing unwanted column and replacing missing values</p>
<p> 2. Combining all 5 weather station files for a month into one file</p>
<p> 3. Built and calling reusable function 2 for cleaning and processing the data such as index resetting, removing unwanted column and grouping the data day-wise by taking their mean values</p>
<p> 4. Combining all 4 months files into one file as 2019 dataset</p>
<p> 5. Analyzing the final 2019 weather dataset</p>

```python
#reusable preprocessing function1
#set index as Date/Time
#removing unnecessary features
#fill missing values with mean via SimpleImputer
from sklearn.impute import SimpleImputer
def preProcessing(df):
    df = df.set_index(pd.DatetimeIndex(df['Date/Time']))
    df = df.drop(['Temp Flag', 'Dew Point Temp Flag', 'Rel Hum Flag', 'Wind Dir Flag', 'Wind Spd Flag', 'Visibility Flag', 'Stn Press Flag', 'Hmdx Flag', 'Wind Chill Flag', 'Visibility (km)', 'Longitude (x)', 'Wind Dir (10s deg)', 'Latitude (y)', 'Stn Press (kPa)', 'Station Name', 'Climate ID', 'Weather','Hmdx','Wind Chill'], axis=1)
    imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
    imputer.fit(df.iloc[:, 5:10])
    df.iloc[:, 5:10] = imputer.transform(df.iloc[:, 5:10])
    return df
```

```python
#Calling preprocessing function1

#dec
df_2019_dec_STANFIELD_1 = preProcessing(df_2019_dec_STANFIELD_1)
df_2019_dec_STANFIELD_2 = preProcessing(df_2019_dec_STANFIELD_2)
df_2019_dec_KOOTENAY = preProcessing(df_2019_dec_KOOTENAY)
df_2019_dec_DOCKYARD = preProcessing(df_2019_dec_DOCKYARD)
df_2019_dec_WINDSOR = preProcessing(df_2019_dec_WINDSOR)
#nov
df_2019_nov_STANFIELD_1 = preProcessing(df_2019_nov_STANFIELD_1)
df_2019_nov_STANFIELD_2 = preProcessing(df_2019_nov_STANFIELD_2)
df_2019_nov_KOOTENAY = preProcessing(df_2019_nov_KOOTENAY)
df_2019_nov_DOCKYARD = preProcessing(df_2019_nov_DOCKYARD)
df_2019_nov_WINDSOR = preProcessing(df_2019_nov_WINDSOR)
#oct
df_2019_oct_STANFIELD_1 = preProcessing(df_2019_oct_STANFIELD_1)
df_2019_oct_STANFIELD_2 = preProcessing(df_2019_oct_STANFIELD_2)
df_2019_oct_KOOTENAY = preProcessing(df_2019_oct_KOOTENAY)
df_2019_oct_DOCKYARD = preProcessing(df_2019_oct_DOCKYARD)
df_2019_oct_WINDSOR = preProcessing(df_2019_oct_WINDSOR)
#sep
df_2019_sep_STANFIELD_1 = preProcessing(df_2019_sep_STANFIELD_1)
df_2019_sep_STANFIELD_2 = preProcessing(df_2019_sep_STANFIELD_2)
df_2019_sep_KOOTENAY = preProcessing(df_2019_sep_KOOTENAY)
df_2019_sep_DOCKYARD = preProcessing(df_2019_sep_DOCKYARD)
df_2019_sep_WINDSOR = preProcessing(df_2019_sep_WINDSOR)
```

```python
#Combining all the monthly station files (grouped by Date/Time 2019)

df_2019_dec = pd.concat([df_2019_dec_STANFIELD_1, df_2019_dec_STANFIELD_2, df_2019_dec_KOOTENAY, df_2019_dec_DOCKYARD, df_2019_dec_WINDSOR]).groupby(df_2019_dec_STANFIELD_1['Date/Time']).mean()
df_2019_nov = pd.concat([df_2019_nov_STANFIELD_1, df_2019_nov_STANFIELD_2, df_2019_nov_KOOTENAY, df_2019_nov_DOCKYARD, df_2019_nov_WINDSOR]).groupby(df_2019_nov_STANFIELD_1['Date/Time']).mean()
df_2019_sep = pd.concat([df_2019_sep_STANFIELD_1, df_2019_sep_STANFIELD_2, df_2019_sep_KOOTENAY, df_2019_sep_DOCKYARD, df_2019_sep_WINDSOR]).groupby(df_2019_sep_STANFIELD_1['Date/Time']).mean()
df_2019_oct = pd.concat([df_2019_oct_STANFIELD_1, df_2019_oct_STANFIELD_2, df_2019_oct_KOOTENAY, df_2019_oct_DOCKYARD, df_2019_oct_WINDSOR]).groupby(df_2019_oct_STANFIELD_1['Date/Time']).mean()
```

```python
#reusable preprocessing function2
#reseting Date/Time index
#drop Date/Time and grouping day-wise by taking their mean

def preProcessing2(df):
    df = df.reset_index()
    df = df.drop(['Date/Time'], axis=1)
    df = df.groupby('Day').mean()
    return df
```

```python
#Calling preprocessing function2
df_2019_dec = preProcessing2(df_2019_dec)
df_2019_sep = preProcessing2(df_2019_sep)
df_2019_nov = preProcessing2(df_2019_nov)
df_2019_oct = preProcessing2(df_2019_oct)
```

```python
#Concatenating the four month data into one file
frames = [df_2019_sep, df_2019_oct, df_2019_nov, df_2019_dec]
df_2019_public_weather_datasets = pd.concat(frames)
```
